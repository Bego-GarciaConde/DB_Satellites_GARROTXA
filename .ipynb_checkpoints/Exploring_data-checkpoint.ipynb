```python
import mysql.connector
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#RAW connection
db = mysql.connector.connect(
        host= "localhost",
        user = "bego",
        passwd = "prueba123",
        database = "Rockstar_satellites_GARROTXA"
    
)
mycursor = db.cursor()
```

```python
# Execute your SQL query
selectquery =  """SELECT id
                FROM satellites_characteristics"""
mycursor.execute(selectquery)

# Fetch the results
records = mycursor.fetchall()

# Convert the results to a DataFrame
column_names = [desc[0] for desc in mycursor.description]
df = pd.DataFrame(records, columns=column_names)
```

```python
for table_name in np.array(df["id"]):
    mycursor.execute(            
       f"""DROP TABLE H{table_name};""")
    db.commit()
```

```python
for table_name in np.array(df["id"]):
    mycursor.execute(
    f"""CREATE TABLE DM{table_name}(
            Snapshot INT PRIMARY KEY,
            Lookback DECIMAL (5, 2),
            X DECIMAL (5, 2),
            Y DECIMAL (5, 2),
            Z DECIMAL (5, 2),
            R DECIMAL (5, 2),
            Mass DECIMAL (12, 2),
            Mstars DECIMAL (12, 2)); """)
    db.commit()
    mycursor.execute(            
           f""" LOAD DATA INFILE '/var/lib/mysql-files/satellites_tables/{table_name}.csv'
                    INTO TABLE DM{table_name}
                    FIELDS TERMINATED BY ','
                    OPTIONALLY ENCLOSED BY '"'
                    LINES TERMINATED BY '\n'
                    IGNORE 1 ROWS
                    (@dummy, @Snapshot, Lookback, @X, @Y, @Z, @R, @Mass, @Mstars)
                    SET Snapshot = NULLIF(@Snapshot, ''),
                    X = NULLIF(@X,''),
                    Y = NULLIF(@Y,''),
                    Z = NULLIF(@Z,''),
                    R = NULLIF(@R,''),
                    Mass = NULLIF(@XMass,''),
                    Mstars = NULLIF(@Mstars,'')""")
    db.commit()
        
        
```

```python
# Execute your SQL query
selectquery =  """SELECT id, M200, Mstars
                FROM satellites_characteristics;"""
mycursor.execute(selectquery)

# Fetch the results
records = mycursor.fetchall()

# Convert the results to a DataFrame
column_names = [desc[0] for desc in mycursor.description]
df = pd.DataFrame(records, columns=column_names)
```

```python
bins = np.arange(7,10, 0.5)
print(bins)
df["log_M200"] = np.log10(df["M200"])
df["log_Mstars"] = np.log10(df["Mstars"])
fig = plt.figure()
#age_bins =np.arange(,12,0.5)
a=plt.hist(np.array(df["log_M200"]), alpha = 0.7, label= "Satellites", bins = bins,rwidth=0.95)
a=plt.hist(np.array(df["log_Mstars"]), alpha = 0.7, label= "Satellites (Luminous)", bins = bins,rwidth=0.95)
plt.xlabel("log Mass")
plt.ylabel("Number of satellites")
plt.legend()
#plt.ylim(0,17)
plt.grid(axis='y', alpha=0.75)
```

```python
selectquery =  """SELECT id, M200, Rmin
                FROM satellites_characteristics
                WHERE M200 > 5E8 AND Rmin < 50 ORDER BY M200 DESC;"""
mycursor.execute(selectquery)

# Fetch the results
records = mycursor.fetchall()

# Convert the results to a DataFrame
column_names = [desc[0] for desc in mycursor.description]
df = pd.DataFrame(records, columns=column_names)
fig, ax = plt.subplots(figsize = (10,4))

for id in df["id"]:
    selectquery =  f"""SELECT Lookback, R
                FROM DM{id};"""
    mycursor.execute(selectquery)
    
    # Fetch the results
    records = mycursor.fetchall()
    
    # Convert the results to a DataFrame
    column_names = [desc[0] for desc in mycursor.description]
    df_halo = pd.DataFrame(records, columns=column_names)
    ax.plot(df_halo["Lookback"],df_halo["R"],label = f"DM{id}",ls  = "-.", alpha = 0.9)
    

ax.set_xlim(6.3,0)
ax.set_ylim(5,120)
ax.grid()
ax.legend(bbox_to_anchor = (1., 1),ncol=2)
ax.set_xlabel(r't$_{lookback}$ [Gyr]', fontsize = 12)
ax.set_ylabel('Distance', fontsize = 12)
```
